```python
# ---------------------------------------------------------------------------------------------------------------------- #
#                     THIS PROGRAM WILL TELL YOU ALL THE SONGS AN ARTIST HAS HAD ON THE CHARTS                           #
# ---------------------------------------------------------------------------------------------------------------------- #
```

```python
import pandas as pd

df = pd.read_csv(r'Hot Stuff.csv')
df = df[['Song', 'Performer', 'Peak Position', 'Weeks on Chart']]
```

```python
def find_artist_hits(artist, df): 
# filtering so that you only have songs by the specified artist
    songs = df.loc[df['Performer'].str.contains(artist)]
    songs.reset_index(drop=True, inplace=True)
    songs = songs.groupby(['Song', 'Performer']).agg({'Peak Position': 'min', 'Weeks on Chart': 'max'})
# if the artist is not on the charts, it will just print 'Artist not found.'
    if songs.empty == True:
        print('Artist not found.')

    return songs
```

```python
artist = input('Enter an artist: ')
# printing all the songs by the specified artist and sorting them by peak rank
find_artist_hits(artist, df).sort_values('Peak Position')
```
